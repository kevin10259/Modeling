```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 200, endpoint=False)
x = np.sin(2 * np.pi * 25 * t)
t2= np.linspace(0, 1, 200)


plt.plot(t, x,'black' ,label= 'Correct Signal',)
plt.plot(t2, np.sin(2 * np.pi * 10 * t2), 'r', label = 'Alias')

plt.title('Sine Wave Aliasing')
plt.xlabel('Frequency Hz')
plt.ylabel('Amplitude')
plt.legend()
```

```python
time= np.arange(0, 500, 6)
amplitude= np.sin(-time)
plt.plot(time, amplitude)

plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
Fs = 500
f = 10
sample = 500
x = np.arange(sample)
y = np.sin(2 * np.pi * f * x / Fs)
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.show()
```
