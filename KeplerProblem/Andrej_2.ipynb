```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
#from scipy import stats
#from scipy import signal
from scipy import integrate
import math
from matplotlib.animation import FuncAnimation

pi = math.pi #pi
P = 3963 #3962.86 #weeks
tp = 0
t = 0 #time steps in weeks
e= 0.967
rp = 0.587 #AU
a = 17.78788 #AU
ra = 34.98876 #AU

#1
M = []
while t < 3963:
    M.append(2*pi*(((t-tp)%P)/P))
    t = t+1
stand = [i for i in range(3963)]
plt.scatter(stand, M)
plt.title('Mean Anomaly')
plt.xlabel('Time')
plt.ylabel('Radians')
```

```python
n=0
En2=np.array([])
t=0   
while t<3963:
    n=0
    En = M[t]
    while n < 100:
        En1 = En-((En-(e*math.sin(En))-M[t])/(1-(e*math.cos(En))))
        En = En1
        n = n+1
    En2 = np.append(En2, En1)
    t=t+1
plt.scatter(stand,En2 )
plt.title('Eccentric Anomaly')
plt.xlabel('Time')
plt.ylabel('Radians')
```

```python
def rk4(f, x0, t):
    n = len(t)
    x = np.array( [ x0 ] * n )
    for i in range( n - 1 ):
        h = t[i+1] - t[i]
        k1 = h * f( x[i], t[i] )
        k2 = h * f( x[i] + 0.5 * k1, t[i] + 0.5 * h )
        k3 = h * f( x[i] + 0.5 * k2, t[i] + 0.5 * h )
        k4 = h * f( x[i] + k3, t[i+1] )
        x[i+1] = x[i] + ( k1 + 2.0 * ( k2 + k3 ) + k4 ) / 6.0

    return x

max_speed = 54.530
M = 1.9891 * (10**30)
G = 6.67408 * (10**(-11))
x_0_au = -0.587
x_0 = x_0_au * 1.496 * (10**(11))
y_0 = 0
v_x_0 = 0
v_y_0 = max_speed * 1000
s0 = np.array([x_0, y_0, v_x_0, v_y_0])
P_rk = 76
amount_orbits = 1000
t_b = P_rk*3.154*(10**(7)) * amount_orbits

def F(s, t):
    return np.array([s[2], s[3], -G*M*s[0]/((s[0]**2 + s[1]**2)**(3/2)), -G*M*s[1] / ((s[0]**2 + s[1]**2)**(3/2))])
steps = int((100000/10) * amount_orbits)
t_RK = np.linspace(0,t_b,steps)

sol = rk4(F, s0, t_RK)

x_rk = sol[:,0]
y_rk = sol[:,1]
v_x_rk = sol[:,2]
v_y_rk = sol[:,3]

xrk = x_rk * (6.684 * 10**-12)
yrk = y_rk * (6.684 * 10 **-12)

plt.figure(1)
plt.title('R.K. 4th Order Method H.C Orbit')
plt.xlabel('AU')
plt.ylabel('AU')
plt.plot(xrk, yrk)
plt.plot(0,0,'o', label = 'Sun')
plt.legend()
```

```python
import numpy as np
from scipy.integrate import odeint
def twoBody(y, t, mu):
    """
    Two Body function returns the derivative of the state space variables.
INPUTS:
    --- t ---
        A scalar time value.

    --- y ---
        A 6x1 array of the state space of a particle in 3D space
OUTPUTS:
    --- ydot ---
        The derivative of y for the two-body problem

"""

    r = np.sqrt(y[0]**2 + y[1]**2 + y[2]**2)

    ydot = np.empty((6,))

    ydot[0] = y[3]
    ydot[1] = y[4]
    ydot[2] = y[5]
    ydot[3] = (-mu/(r**3))*y[0]
    ydot[4] = (-mu/(r**3))*y[1]
    ydot[5] = (-mu/(r**3))*y[2]

    return ydot


# In m and m/s
# first three are the (x, y, z) position
# second three are the velocities in those same directions respectively
Y0 = np.array([-5614924.5443320004,
               -2014046.755686,
               2471050.0114869997,
               -673.03650300000004,
               582.41158099999996,
               1247.7034980000001])

mu = 3.986004418 * 10**14

solution = odeint(twoBody, Y0, np.linspace(0., 351., 100), args=(mu, ))
```
