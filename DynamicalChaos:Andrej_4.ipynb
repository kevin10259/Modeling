```python
import matplotlib.pyplot as plt

def logistic (R, x0, N):    
    x = x0
    x_list_2 = [x0]
    for i in range(N-1):      
        x = R * x * (1. - x)
        x_list_2.append(x)
    return x_list_2

x_result = logistic (3.15, 0.5, 30)


plt.plot(x_result, 'o:g')
plt.title('Logistic Equation when R = 3.15')
plt.xlabel('Number of Iterations')
plt.ylabel('Initial X')
```

```python
def is_stable(re,im, num_iterations):
    z = 0.0j
    c = complex(re,im)
    for i in range(num_iterations):
        z = z * z + c
        if (z.real**2 + z.imag**2) >= 2:
            return i
    return num_iterations

row = 2000
column = 2000
array = np.zeros([row, column])
for row_index, re in enumerate(np.linspace(-2,1,row)):
    for column_index, im in enumerate(np.linspace(-1,1,column)):
        array[row_index,column_index] = is_stable(re,im,100)
    
```

```python
plt.imshow(array.T,interpolation = 'bilinear', extent = [-2,1,-1,1], cmap = "plasma")
plt.title('Mandelbrot Set With Differing C Values')
#plt.gca().set_aspect("equal")
#plt.axis("off")
#plt.tight_layout()
#plt.show()
```

```python
import os

from PIL.Image import open as load_pic, new as new_pic


def main(path, iterations, keep_all=False, name="Happiness.png"):
    """
    Params
        path:str
            path to photograph
        iterations:int
            number of iterations to compute
        name:str
            formattable string to use as template for file names
    """
    title = os.path.splitext(os.path.split(path)[1])[0]
    counter = 0
    while counter < iterations:
        with load_pic(path) as image:
            dim = width, height = image.size
            with new_pic(image.mode, dim) as canvas:
                for x in range(width):
                    for y in range(height):
                        nx = (2 * x + y) % width
                        ny = (x + y) % height                        
                        canvas.putpixel((nx, height-ny-1), image.getpixel((x, height-y-1)))

        if counter > 0 and not keep_all:
            os.remove(path)
        counter += 1
        print(counter, end="\r")
        path = name.format(name=title, index=counter)
        canvas.save(path)

    return canvas



if __name__ == "__main__":
    path = input("Enter the path to an image:\n\t")
    while not os.path.exists(path):
        path = input("Couldn't find your chosen image, please try again:\n\t")
    result = main(path, 50)
    result.show()

    
    

plt.imshow(result)
```

```python
import os
import numpy as np

Pixel = 105
pic = plt.imread('Happiness.png')

def ArnoldsCat(z):
    cool = plt.imread('Happiness.png')
    x,y = np.meshgrid(range(Pixel), range(Pixel))
    xmap = (2 * x + y) % Pixel
    ymap = (x + y) % Pixel
    d = 0
    while d < z:
        cool = cool[ymap,xmap]
        d = d +1
    return cool
```

```python
plt.imshow(ArnoldsCat(40))
plt.title('Resulting Image After 40 Iterations')
```

```python
z = 1
while z < 500:
    cool = ArnoldsCat(z)
    if np.array_equal(pic,cool) == True:
        break
    else:
        z = z+1
print(z)
```
