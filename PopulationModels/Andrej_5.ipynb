```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import integrate



data = np.genfromtxt('Ecuador_Pop.txt')
P0 = [500]


time = np.linspace(0,300,301)
time_array = [0,300]
k = 0.1
N = 300000000

s0 = P0

def f(t,s):
    P = s
    return [k*P*(1 - (P/N))]

sol_2 = scipy.integrate.solve_ivp(fun = f, t_span = time_array, y0 = s0,t_eval = time)

data_2 = sol_2.y[0]

plt.plot(time, data_2 * 10**-6)
```

```python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy.integrate import odeint

a,b,c,d = 1,0.2,0.15,1.2
F = 10
R = 10

x = [R,F]

def RF_find(t0, x, a0=a, b0=b, c0=c, d0=d):
    r_dot = a0*x[0] - b0*x[0]*x[1]
    f_dot = c0*x[0]*x[1] - d0*x[1]
    return np.array([r_dot, f_dot])


t = np.linspace(0,250,10000) #time range
t_arr = [0,250]


sol_k = scipy.integrate.solve_ivp(fun = RF_find, t_span = t_arr, y0 = x,t_eval = t, rtol = 0.0000001, atol = 1e-8)

sol_rabbit = sol_k.y[0]
sol_fox = sol_k.y[1]

#print(RF_found)

#fig, ax = plt.subplots()
#plt.scatter(RF_found[:,0], RF_found[:,1], label="Rabits v. Foxes")
plt.plot(t, sol_rabbit,  label="Rabbits")
plt.plot(t, sol_fox, label="Foxes")
    

plt.title("Rabbits and Foxes Lotka-Voltera")

plt.legend()
plt.xlabel("Time")
plt.ylabel("Population")
plt.show()

plt.figure()
plt.plot(sol_rabbit,sol_fox)
plt.xlabel('Rabbits')
plt.ylabel('Foxes')
plt.title('Phase Diagram Prey vs Predator')
```

```python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy.integrate import odeint

cont,heal = 0.5,2/7 #contact rate #chance to heal in days
H,S,I = 0.87,0.03,0.10


def virus(x, t0, cont0=cont, heal0=heal):
    H_f = -cont*x[0]*x[1]
    S_f = cont*x[0]*x[1] - heal*x[1]
    I_f = heal*x[1]
    return H_f, S_f, I_f


t = np.linspace(0,100,1000) #time range
x = [H,S,I]

Virus_f = odeint(virus, x, t)
print(x)

H_p,S_p,I_p=Virus_f.T

max_sick = max(Virus_f[:,1])


fig, ax = plt.subplots()
ax.plot(t, H_p, label="Healthy")
ax.plot(t, S_p, label="Sick")
ax.plot(t, I_p, label="Immune")
plt.title('Pandemic Population W/ Inoculation')
ax.legend()
ax.set_xlabel("Time (days)")
ax.set_ylabel("People (%)")

plt.show()

print(max_sick)
```
